```python
# Task1: Variables and Data Types
name = "Anantha"
age = 22
height = 5.9
is_student = True

print(name, age, height, is_student)

# Task2: Sting Formatting
print(f"Name:{name } | Age: {age} | Height: {height} | Student: {is_student}")

# Task3: Arithmetic Operations

# Age in months
age_months = age * 12
print("Age in months:", age_months)

# Age in days
age_days = age * 365
print("Age in days:", age_days)

# Remainder when age divided by 7
remainder = age % 7
print("Remainder when age divided by 7:", remainder)

# Age squared
age_squared = age ** 2
print("Age squared:", age_squared)
```
